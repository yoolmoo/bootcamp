```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(X_b, y, theta, learning_rate, n_iterations):
    for iteration in range(n_iterations):
        gradients = -1 / len(X_b) * X_b.T.dot(y - X_b.dot(theta))
        theta -= learning_rate * gradients
    return theta

# Generate synthetic data for demonstration
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term to the feature matrix
X_b = np.c_[np.ones((100, 1)), X]

# Initialize random parameters (theta)
theta = np.random.randn(2, 1)

# Set hyperparameters
learning_rate = 0.01
n_iterations = 1000

# Perform gradient descent using the function
theta = gradient_descent(X_b, y, theta, learning_rate, n_iterations)

# Display the computed parameters
print("Theta (Parameters) using Gradient Descent:")
print(theta)

# Make predictions on new data
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)

# Visualize the linear regression line
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_new, y_predict, color='red', label='Linear Regression Line (Gradient Descent)')
plt.title('Linear Regression with Gradient Descent')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
